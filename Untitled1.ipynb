```python
import pandas as pd
```

```python
df = pd.read_csv('crime_data.csv')
df.head()
```

```python
df['Risk'] = ""
```

```python
df.head()
```

```python
substring = 'TOTAL'
filter = df['district'].str.contains(substring)
filtered_df = df[~filter]
df = filtered_df
```

```python
df
```

```python
df.loc[df['total_ipc_crimes'] > 3100, 'risk'] = 1
df.loc[df['total_ipc_crimes'] <= 3100, 'risk'] = 0
```

```python
df
```

```python
# Assuming df is your DataFrame
# Initialize counts for 0s and 1s
count_0 = 0
count_1 = 0

# Iterate over columns
for column in df.columns:
    if 'risk' in column:
        # If 'risk' is in the column name, count the occurrences of 0 and 1
        count_0 += df[column].eq(0).sum()
        count_1 += df[column].eq(1).sum()

# Print the counts
print("Number of 1s in 'risk' column:", count_1)
print("Number of 0s in 'risk' column:", count_0)
```

```python
df.to_csv('crime.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('crime.csv')

# 1. Bar Plot: Total IPC Crimes by District
plt.figure(figsize=(12, 6))
plt.bar(data['district'], data['total_ipc_crimes'], color='skyblue')
plt.xlabel('District')
plt.ylabel('Total IPC Crimes')
plt.title('Total IPC Crimes by District')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# 2. Line Plot: Trends of specific crime types over the years for a district (Example: Murder)
district_data = data[data['district'] == 'ADILABAD']  # Change the district as needed
plt.figure(figsize=(10, 6))
plt.plot(district_data['year'], district_data['murder'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Number of Murders')
plt.title('Trends of Murders over Years in Adilabad District')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Scatter Plot: Relationship between two variables (e.g., murders vs. thefts)
plt.figure(figsize=(8, 6))
plt.scatter(data['murder'], data['theft'], color='orange')
plt.xlabel('Number of Murders')
plt.ylabel('Number of Thefts')
plt.title('Relationship between Murders and Thefts')
plt.grid(True)
plt.tight_layout()
plt.show()

# 4. Pie Chart: Distribution of different types of crimes within a district (Example: ADILABAD)
crime_labels = ['Murder', 'Attempt to Murder', 'Rape', 'Kidnapping/Abduction', 'Theft', 'Riots']  # Add more as needed
crime_values = district_data.loc[0, ['murder', 'attempt_to_murder', 'rape', 'kidnapping_abduction', 'theft', 'riots']].values
plt.figure(figsize=(8, 8))
plt.pie(crime_values, labels=crime_labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Crimes in Adilabad District')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```
